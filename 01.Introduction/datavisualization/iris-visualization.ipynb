```python
# Visualization of data

import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import radviz
```

### Basic Iris flower Data set

```python
sns.set(style="white", color_codes=True)

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv("../../data/csv/iris.csv", names=names) # the iris dataset is now a Pandas DataFrame

# Head
print iris.head()
```

```python
print iris["class"].value_counts()
```

# Pandas Plot

## Normal Plot with selected columns

```python
iris.plot(kind="scatter", x="sepal-length", y="sepal-width")

plt.show()
```

```python
iris.plot(kind="scatter", x="petal-length", y="petal-width")

plt.show()
```

### Seaborn Pairplot with Various Diagram Kinds

```python
diagkinds = ['auto', 'hist', 'kde']

for dk in diagkinds:

    sns.pairplot(iris, hue="class", size=3, diag_kind=dk)
    plt.show()
    print '\n----------------------------\n'
```

### Radviz Plot

```python
radviz(iris, "class")
plt.show()
```

### Box Plot

```python
iris.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
plt.show()
```

### Histogram Plot

```python
iris.hist()
plt.show()
```

```python
correlations = iris.corr()
print correlations
```

```python
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ax.set_xticklabels(['','sepal-length', 'sepal-width', 'petal-length', 'petal-width'])
ax.set_yticklabels(['','sepal-length', 'sepal-width', 'petal-length', 'petal-width'])
plt.show()
```

```python
for n in names:
    if n != 'class':
        sns.violinplot(x="class", y=n, data=iris, size=6)
        plt.show()
```
